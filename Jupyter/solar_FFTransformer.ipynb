```python
def visual(true, preds=None, label1="GroundTruth", label2= "Prediction", name='./pic/test.pdf'):
    """
    Results visualization
    """
    plt.figure()
    plt.plot(true, label=label1, linewidth=2)
    if preds is not None:
        plt.plot(preds, label=label2, linewidth=2)
    plt.legend()
    # plt.savefig(name, bbox_inches='tight')
    plt.show()
```
